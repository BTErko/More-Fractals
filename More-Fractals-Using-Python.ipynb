```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<font size='3'> The first function is used to translate, store and plot the specific pattern of the fractal. I will test the above function by plotting a simple geometric shape, the triangle. The triangle pattern and rule is F--F--F with an angle of rotation 60 degrees or F-F-F with an angle of rotation of 120 degrees. As a reminder, F stands for move forward and G is used for move vertical. The length of movement is set to 1 unit and the starting position is set to (0, 0).</font> <br><br>

```python
# Function to draw fractals

def drawFractal(instruction, angle, x0, y0):
    x = []
    y = []

    angChange = {'+': -1*angle, '-': angle} # dictionary for angle of rotation
    
    x.append(0)
    y.append(0)
    l,x0,y0,angle = 1,0,0,0   # initialize lenth of movement, angle and starting position 
    for i in range(1,len(instruction)+1):
        if(instruction[i-1] == 'F' or instruction[i-1] == 'G' or instruction[i-1] == 'X' or instruction[i-1] == 'Y'):
            dx = x[i-1] + l * np.cos(angle*np.pi/180)
            x.append(dx)
            dy = y[i-1] + l * np.sin(angle*np.pi/180)
            y.append(dy)
     
        elif(instruction[i-1] == '+'):
            angle += angChange['+'] 
            x.append(x[i-1])
            y.append(y[i-1])
            
        elif(instruction[i-1] == '-'):
            angle += angChange['-'] 
            x.append(x[i-1])
            y.append(y[i-1])


    figure, axis = plt.subplots(figsize= (8,8))
    axis.plot(x,y, color = 'darkmagenta')
    plt.axis('equal')
    plt.show()
```

<center> <font size='3'> <b> Quadratic Koch Island </b> </font> <br>

<font size='3'>The quadratic Kock Island starts with F+F+F+F and the rule is F -> F+F-F-FFF+F+F-F with an angle of rotation 90 degrees. </font> <br><br>

```python
def quadraticKochIsland(iteration, pattern):
    if(iteration == 0):
        pattern = pattern
    else:
        for i in range(iteration):
            pattern = pattern.replace('F', 'F+F-F-FFF+F+F-F')
    return pattern
```

```python
drawFractal(quadraticKochIsland(5, 'F+F+F+F'), 90, 0, 0)
```

<center> <font size='3'> <b> Quadratic Snowflake </b> </font> <br>

<font size='3'>The Quadratic Snowflake starts with F and the rule is F -> F-F+F+F-F
 with an angle of rotation 90 degrees. </font> <br><br>

```python
def quadraticSnowflake(iteration, pattern):
    if (iteration == 0):
        pattern = pattern
    else:
        for i in range(iteration):
            pattern = pattern.replace('F', 'F-F+F+F-F')
    return pattern
    
```

```python
drawFractal(quadraticSnowflake(8, 'F'), 90, 0, 0)
```

<center> <font size='3'> <b> Levy C Curve </b> </font> <br>

```python
def levyCCurve(iteration, pattern):
    if (iteration == 0):
        pattern = pattern
    else:
        for i in range(iteration):
            pattern = pattern.replace('F', '+FX--F+')
            
    return pattern
```

```python
drawFractal(levyCCurve(10, 'F'), 45, 0, 0)
```

<center> <font size='3'> <b> Crystal </b> </font> <br>

```python
def crystal(iteration, pattern):
    if (iteration == 0):
        pattern = pattern
    else:
        for i in range(iteration):
            pattern = pattern.replace('F', 'FF+F++F+F')
            
    return pattern
```

```python
drawFractal(crystal(5, 'F+F+F+F'), 90, 0, 0)
```
